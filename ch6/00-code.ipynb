```python
from matplotlib import pyplot as plt
import numpy as np

from phaseportrait import PhasePortrait2D
```

```python
# 6.1.10
def f(x, y):
    return (
        y + y**2,
        -x/2 + y/5 - x*y + 6/5*y**2 
    )

example1 = PhasePortrait2D(f, [-15, 10], MeshDim=40)
#example1.add_nullclines(xcolor="black", ycolor="green")
fig, ax = example1.plot()
```

```python
# 6.3.1
A = np.array([
    [1, -1],
    [-4, 0]
])
np.linalg.eig(A)
# This tells us that (-2, -2) is a saddle point
```

```python
def dF(x, y):
    return (
        x-y,
        x**2 - 4
    )
```

```python
dF(-3, 0)
```

```python
phase_portrait = PhasePortrait2D(dF, [-10, 10], MeshDim=40)
#example1.add_nullclines(xcolor="black", ycolor="green")
fig, ax = phase_portrait.plot()
```

```python
# 6.5.14
def get_dF(D):
    def dF(v, theta):
        return (
            -np.sin(theta) - D*v**2,
            (-np.cos(theta) + v**2) / v
        )
    return dF
```

```python
phase_portrait = PhasePortrait2D(get_dF(D=0), [-3, 3], MeshDim=100, xlabel="v", ylabel="theta")
#example1.add_nullclines(xcolor="black", ycolor="green")
fig, ax = phase_portrait.plot()
```

```python
# What if D != 0
phase_portrait = PhasePortrait2D(get_dF(D=1), [-2*np.pi, 2*np.pi], MeshDim=40, xlabel="v", ylabel="theta")
fig, ax = phase_portrait.plot()
```
